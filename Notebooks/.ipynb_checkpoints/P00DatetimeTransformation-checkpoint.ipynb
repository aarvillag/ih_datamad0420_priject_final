```python
# Notebook for date-time formating from original dataset
# We will create two additional columns with datetime in US and EU format
```

```python
import pandas as pd
```

```python
df_es=pd.read_csv('ES.csv', header=0, names=['date','time','open','max','min','close','volume'])
df_es.head(10)
```

```python
df_es.info()
```

```python
df_es['datetime_US']=pd.to_datetime(df_es[['date', 'time']].agg(' '.join, axis=1), format='%m/%d/%Y %H:%M')
df_es['datetime_EU']= df_es['datetime_US'] + pd.to_timedelta('6 h')

#df_es.drop('time_US', axis=1, inplace=True)
```
